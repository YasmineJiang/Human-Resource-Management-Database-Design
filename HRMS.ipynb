```python
#Q7:Fetch First name, last name, middle name, Country name, Designation name and DOB within a range 
#(e.g.: 1/2/1984 to 1/3/1986).
query_7 = "SELECT Emp_First_Name, Emp_Middle_Name, Emp_Last_Name, Emp_DOB, Country_Name, Desig_Name FROM Employee_Details e LEFT JOIN Country c ON e.Emp_Country_id = c.Country_id LEFT JOIN Designation d ON e.Desig_id = d.Desig_id WHERE Emp_DOB BETWEEN '1984-02-01' AND '1986-03-01'"
df_7 = pd.read_sql_query(query_7, engine)
df_7
```

```python
#Q8:Fetch First name, Last name, middle name, Country name, Designation name and his current salary.
query_8 = """
WITH T1 AS (
    SELECT Emp_First_Name, Emp_Last_Name, Country_Name, Desig_Name, Emp_Salary_Change_Year, Emp_Salary, MAX(Emp_Salary_Change_Year) OVER (PARTITION BY s.Emp_id) 'Current'
    FROM Employee_Details e
    LEFT JOIN Country c 
    ON e.Emp_Country_id = c.Country_id 
    LEFT JOIN Designation d 
    ON e.Desig_id = d.Desig_id 
    LEFT JOIN Salary s 
    ON e.Emp_id = s.Emp_id
) 

SELECT Emp_First_Name, Emp_Last_Name, Country_Name, Desig_Name, Emp_Salary
FROM T1
WHERE Emp_Salary_Change_Year = Current;
"""

df_8 = pd.read_sql_query(query_8, engine)
df_8
```

```python
#Q9:Fetch First name, Last name, middle name, Country name, Designation name and his/her highest salary within a range 
#(e.g. : 50000 to 100000).
query_9 = """
WITH T1 AS (
    SELECT Emp_First_Name, Emp_Last_Name, Country_Name, Desig_Name, Emp_Salary, MAX(Emp_Salary) OVER (PARTITION BY s.Emp_id) 'Current'
    FROM Employee_Details e
    LEFT JOIN Country c 
    ON e.Emp_Country_id = c.Country_id 
    LEFT JOIN Designation d 
    ON e.Desig_id = d.Desig_id 
    LEFT JOIN Salary s 
    ON e.Emp_id = s.Emp_id
) 

SELECT Emp_First_Name, Emp_Last_Name, Country_Name, Desig_Name, Emp_Salary
FROM T1
WHERE Emp_Salary = Current;
"""

df_9 = pd.read_sql_query(query_9, engine)
df_9
```

```python
#Q13:Get Designation, total salary spend by the company with respect to each 
query_13 = """
    SELECT Desig_Name, SUM(Emp_Salary) AS Total_Salary
    FROM Employee_Details e
    LEFT JOIN Designation d 
    ON e.Desig_id = d.Desig_id 
    LEFT JOIN Salary s 
    ON e.Emp_id = s.Emp_id
    GROUP BY Desig_Name
"""

df_13 = pd.read_sql_query(query_13, engine)
df_13
```

```python
#Q14:Get designation name, number of employees in a designation with respect to a designation from employee details table.
query_14 = """
    SELECT Desig_Name, COUNT(DISTINCT Emp_Id) AS Num_of_Emp
    FROM Employee_Details e
    LEFT JOIN Designation d 
    ON e.Desig_id = d.Desig_id 
    GROUP BY Desig_Name
"""

df_14 = pd.read_sql_query(query_14, engine)
df_14
```

```python
#Q15:Get designation wise average salaries from employee details table and order by Salary ascending.
query_15 = """
    SELECT Desig_Name, AVG(Emp_Salary) AS Total_Salary
    FROM Employee_Details e
    LEFT JOIN Designation d 
    ON e.Desig_id = d.Desig_id 
    LEFT JOIN Salary s 
    ON e.Emp_id = s.Emp_id
    GROUP BY Desig_Name
    ORDER BY Total_Salary
"""

df_15 = pd.read_sql_query(query_15, engine)
df_15
```

```python
#Q16:Select no of employees joined with respect to year and month from Employee details table.mployee_Details"
query_16 = """
    SELECT YEAR(Emp_JoinDate) AS Joined_year, MONTH(Emp_JoinDate) AS Joined_month, COUNT(DISTINCT Emp_Id) AS Num_of_Emp
    FROM Employee_Details e
    GROUP BY Joined_year, Joined_month
"""
df_16 = pd.read_sql_query(query_16, engine)
df_16
```

```python
#Q19:Select the employee with highest appraisal. (difference of first salary and current salary)
query_19 = """
WITH T1 AS (SELECT Emp_Id, 
    MAX(Emp_Salary) OVER (PARTITION BY Emp_Id) 'current', 
    MIN(Emp_Salary) OVER (PARTITION BY Emp_Id) 'previous'  
    FROM Salary),
    
T2 AS(SELECT DISTINCT Emp_Id, (current - previous) AS appraisal
FROM T1),

T3 AS(SELECT Emp_Id, appraisal,
RANK() OVER (ORDER BY appraisal DESC) 'ranking'
FROM T2)

SELECT Emp_Id
FROM T3
WHERE ranking = 1


"""
df_19 = pd.read_sql_query(query_19, engine)
df_19
```
